```python
dbutils.widgets.text("catalog","dev")
```

```python
catalog = dbutils.widgets.get("catalog")
```

```python
from pyspark.sql.functions import *
```

```python
df_customers = spark.read.csv("/Volumes/dev/bronze/raw/customers.csv", header=True, inferSchema=True)
df_order_items = spark.read.csv("/Volumes/dev/bronze/raw/order_items.csv",header=True,inferSchema=True)
df_orders = spark.read.csv("/Volumes/dev/bronze/raw/orders.csv",header=True,inferSchema=True)
df_products = spark.read.csv("/Volumes/dev/bronze/raw/products.csv",header=True,inferSchema=True)
```

```python
df_customers = df_customers.withColumn("Ingestion_Date",  to_date(current_timestamp())).withColumn("File_Path",col("_metadata.file_path"))
df_order_items = df_order_items.withColumn("Ingestion_Date", to_date(current_timestamp())).withColumn("File_Path",col("_metadata.file_path"))   
df_orders = df_orders.withColumn("Ingestion_Date", to_date(current_timestamp())).withColumn("File_Path",col("_metadata.file_path"))
df_products = df_products.withColumn("Ingestion_Date", to_date(current_timestamp())).withColumn("File_Path",col("_metadata.file_path"))
```

```python
df_customers.write.mode("overwrite").partitionBy("Ingestion_Date").saveAsTable(f"{catalog}.bronze.customers")
df_order_items.write.mode("overwrite").partitionBy("Ingestion_Date").saveAsTable(f"{catalog}.bronze.order_items")
df_orders.write.mode("overwrite").partitionBy("Ingestion_Date").saveAsTable(f"{catalog}.bronze.orders")
df_products.write.mode("overwrite").partitionBy("Ingestion_Date").saveAsTable(f"{catalog}.bronze.products")
```
